```python
import pandas as pd
import pickle
import os
```

```python
for f in os.listdir('../data_vis/tsne_data/'):
    num = f.split(".")[0].split("_")[-1]
    with open(os.path.join("../data_vis/tsne_data/", f), "rb") as f:
        df = pickle.load(f)
        df = df[["X", "Y"]]
    
    with open(f'../data_vis/caption_data/cap_df_{num}.pkl', 'rb') as f:
        df_cap = pickle.load(f)
        df_cap = df_cap[["caption", "mean", "votes"]]
    
    out = pd.concat([df, df_cap], axis=1)

    out.to_json(f'tsne/{num}.json', orient="records")

    
    

```
